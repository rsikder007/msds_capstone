```python
import dash
import pandas as pd

from dash import html
from dash import dcc

import pandas as pd
import plotly.express as px
```

```python
#!pip install dash
```

```python
# Load the data from the CSV file
data = pd.read_csv("charge_st.csv")
```

```python
data
```

```python
# Create a scatter plot using Plotly Express
fig = px.scatter(data, x="chargeTimeHrs", y="kwhTotal")
```

```python
# Create the Dash app
app = dash.Dash(__name__)
```

```python
# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='EV Charging Data'),
    dcc.Graph(
        id='charging-data',
        figure=fig
    )
])
```

```python
# Run the app
if __name__ == '__main__':
    app.run_server(debug=False)
```
